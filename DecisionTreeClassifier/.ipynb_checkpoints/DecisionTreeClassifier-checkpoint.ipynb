```python
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
import sklearn as skl
import graphviz 
from sklearn import tree
```

```python
data = pd.read_csv("train.csv")
# clf = tree.DecisionTreeClassifier()
clf_sample_leaf_1 = tree.DecisionTreeClassifier(min_samples_leaf=1)
clf_sample_leaf_2 = tree.DecisionTreeClassifier(min_samples_leaf=5)
clf_sample_leaf_3 = tree.DecisionTreeClassifier(min_samples_leaf=10)
# training data 
xtrain = data.iloc[0:21000,1:785]
train_label = data.iloc[0:21000,0]

# testing data 
xtest = data.iloc[21000:42000,1:785]
test_label = data.iloc[21000:42000,0]
```

```python
# Training Decision Tree
# clf.fit(xtrain,train_label)
# first model train has accuracy of about 0.838 
# clf.fit(xtrain,train_label)
clf_sample_leaf_1.fit(xtrain,train_label)
clf_sample_leaf_2.fit(xtrain,train_label)
clf_sample_leaf_3.fit(xtrain,train_label)
```

```python
# Computing accuracy
# p = clf.predict(xtest)
p_sample_leaf_1 = clf_sample_leaf_1.predict(xtest)
p_sample_leaf_2 = clf_sample_leaf_2.predict(xtest)
p_sample_leaf_3 = clf_sample_leaf_3.predict(xtest)
# print("General Accuracy score: \n", skl.metrics.classification_report(test_label, p))
# print ("######################################################################")
# print ("######################################################################")
print("General Accuracy score: \n", skl.metrics.classification_report(test_label, p_sample_leaf_1))
print ("######################################################################")
print ("######################################################################")
print("General Accuracy score: \n", skl.metrics.classification_report(test_label, p_sample_leaf_2))
print ("######################################################################")
print ("######################################################################")
print("General Accuracy score: \n", skl.metrics.classification_report(test_label, p_sample_leaf_3))
# print("General Accuracy score: " + skl.metrics.accuracy_score(test_label, p))
# print("General Accuracy score: " + skl.metrics.accuracy_score(test_label, p))

```

```python
#Testing single input value
for testing_value in range(10):
    d = xtest[testing_value]
    d.shape = (28,28)
    pt.imshow(d,cmap='gray')
    print(clf.predict([xtest[testing_value]]))
    pt.show()
```

```python
#Data Visualization/Decision Tree Visualization
dot_data = tree.export_graphviz(clf_sample_leaf_1, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("dot","Digit Recognition - clf_sample_leaf_1")
graph

dot_data = tree.export_graphviz(clf_sample_leaf_2, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("dot","Digit Recognition - clf_sample_leaf_2")
graph

dot_data = tree.export_graphviz(clf_sample_leaf_3, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("dot","Digit Recognition - clf_sample_leaf_3")
graph
```
